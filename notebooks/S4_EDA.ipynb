# TrustChain S4 Week 2 EDA

This notebook verifies the cleaned NYC dataset and the GeoJSON export used by the React map.

Cells:
1. Load the cleaned CSV and inspect shape plus category counts.
2. Load the GeoJSON and inspect the first feature.
3. Render a quick Folium preview of the first 100 venues.

```python
import pandas as pd

df = pd.read_csv('../data/processed/foursquare_nyc_clean.csv')
print(df.shape)
print(df['venue_category'].value_counts())
```

```python
import json

with open('../frontend/src/data/venues.geojson') as f:
    geo = json.load(f)

print(f"Features: {len(geo['features'])}")
print(geo['features'][0])
```

```python
import folium

m = folium.Map(location=[40.73, -73.93], zoom_start=11)
for feat in geo['features'][:100]:
    c = feat['geometry']['coordinates']
    folium.CircleMarker(
        [c[1], c[0]],
        radius=4,
        popup=feat['properties']['category'],
    ).add_to(m)

m.save('nyc_preview.html')
print('Map saved!')
```
